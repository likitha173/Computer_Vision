```python
import cv2
cv2.__version__
from matplotlib import pyplot as plt
```

### Reading image

```python
img=cv2.imread('opencv.png')
name='openCV python image'
plt.title(name)
plt.imshow(img)
```

### Flipping the image Horizontally

```python
horizontal_flip=cv2.flip(img,1)
names='opencv python flip image horizontally'
plt.title(name)
plt.imshow(horizontal_flip)
```

### flipping the image vertically and horizontally

```python
vertical_horizontal_flip=cv2.flip(img,-1)
name='opencv python flip vertically and horizontally'
plt.title(name)
plt.imshow(vertical_horizontal_flip)
```

### basic reading saving displaying impage

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import cv2
image=cv2.imread('baby.jpg')
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.title('original')
plt.imshow(image_rgb)
```

```python
gray_image=cv2.imread('baby.jpg',cv2.COLOR_BGR2GRAY)
## SAVING USING imwrite()
cv2.imwrite('baby_gray.jpeg',gray_image)
plt.title('Gray_scale')
plt.imshow(gray_image)
```

### Loading and display the saved image

```python
#loading
img=cv2.imread('baby_gray.jpeg')
#Display
plt.title('saved image')
plt.imshow(img)
```
