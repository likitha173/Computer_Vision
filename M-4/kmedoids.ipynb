```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn_extra.cluster import KMedoids
import warnings
warnings.filterwarnings("ignore")
```

```python
# Creating Dummy image data
```

```python
data=np.array([[7,8],[9,10],[11,5],[4,9],[7,5],[2,3],[4,5]])
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=data[:,0],y=data[:,1],marker='o',s=100)
plt.title("dummy image data points")
plt.show()
```

```python
center=[[4,5],[9,10]]
```

```python
Kmedoids=KMedoids(n_clusters=2)
Kmedoids.fit(data)

labels=Kmedoids.labels_
```

```python
unique_labels=set(labels)

color_plot=[
    plt.cm.Spectral(each) for each in np.linspace(0,1,len(unique_labels))
]
xy=data

plt.figure(figsize=(10,8))
sns.scatterplot(x=xy[:,0],y=xy[:,1],marker='+',hue=labels,s=100)

sns.scatterplot(Kmedoids.cluster_centers_[:,0],Kmedoids.cluster_centers_[:,1],marker='o',color="red",label="centroids",s=100)
plt.title("kmedoids clustering on dummy data",fontsize=14)
plt.legend()
plt.show()
```
